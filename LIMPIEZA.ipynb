# **Ejemplo procesamiento de at√≠picos en un Data Frame**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
ca = fetch_california_housing()
```

```python
ca
```

```python
dir(ca)
```

```python
X = ca.data
y = ca.target
```

```python
data = np.concatenate((X,y.reshape(-1,1)), axis=1)
```

```python
df = pd.DataFrame(data, columns=ca.feature_names+['target'])
df.head()
```

```python
df.drop(['Latitude', 'Longitude'], axis=1, inplace=True)
```

```python
features = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup',]
features
```

```python
sns.set_style('dark')
for col in features:
    plt.figure(figsize=(15,4))
    plt.subplot(131)
    sns.distplot(df[col], label="skew: " + str(np.round(df[col].skew(),2)))
    plt.legend()
    plt.subplot(132)
    sns.boxplot(df[col])
    plt.subplot(133)
    stats.probplot(df[col], plot=plt)
    plt.tight_layout()
    plt.show()
```

```python
df_cap = df.copy()
```

```python
def percentile_capping(df, cols, from_low_end, from_high_end):
    
    for col in cols:
        
#         lower_bound = df[col].quantile(from_low_end)
#         upper_bound = df[col].quantile(1-from_high_end)
        
#         df[col] = np.where(df[col]>upper_bound, upper_bound,
#                  np.where(df[col]<lower_bound, lower_bound, df[col]))

        stats.mstats.winsorize(a=df[col], limits=(from_low_end, from_high_end), inplace=True)
```

```python
# fourth argument of 0.01 from right end is equivalent to saying
# 1-0.01 = 0.99 quantile from the left end
percentile_capping(df_cap, features, 0.01, 0.01)
```

```python
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Obtener las variables predictoras (X) y la variable dependiente (Y) del DataFrame
X = df_cap[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population','AveOccup']]
Y = df_cap['target']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Crear y entrenar el modelo de regresi√≥n Ridge
ridge = Ridge(alpha=optimal_alpha)  # Par√°metro de regularizaci√≥n (ajustar seg√∫n sea necesario)
ridge.fit(X_train, Y_train)

# Realizar predicciones en el conjunto de prueba
Y_pred = ridge.predict(X_test)

# Calcular el error cuadr√°tico medio (MSE)
mse = mean_squared_error(Y_test, Y_pred)

# Imprimir el MSE
print("Error cuadr√°tico medio (MSE):", mse)
```

Este modelo corrige la multicolinealidad y proporciona el mejor modelo, es decir, el de menor MSE.

## **An√°lisis de los residuos**

```python
# Obtener los residuos
residuos = Y_test - Y_pred
```

```python
# Crear un DataFrame con la variable de residuos
residuos = pd.DataFrame({'residuos': residuos})

# Imprimir el DataFrame
residuos
```

```python
import seaborn as sns

# Visualizar la distribuci√≥n de los residuos
sns.histplot(residuos, kde=True)
```

```python
import matplotlib.pyplot as plt

# Crear el gr√°fico de puntos
plt.scatter(residuos.index, residuos['residuos'])

# Agregar etiquetas y t√≠tulo al gr√°fico
plt.xlabel('√çndice')
plt.ylabel('Residuos')
plt.title('Gr√°fico de Puntos de Residuos')

# Mostrar el gr√°fico
plt.show()
```

```python
residuos.describe()
```

La media de los residuos es aproximadamente cero.

```python
import scipy.stats as stats

# Realizar el contraste de hip√≥tesis
residuos = residuos['residuos']  # Extraer la columna de residuos del DataFrame

# Realizar el test t de una muestra
t_statistic, p_value = stats.ttest_1samp(residuos, 0)

# Imprimir los resultados del contraste
print('Estad√≠stico t:', t_statistic)
print('Valor p:', p_value)
```

Se acepta que la media de los residuos es cero.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Calcular la funci√≥n de autocorrelaci√≥n simple
plot_acf(residuos['residuos'], lags=10)
plt.xlabel('Lags')
plt.ylabel('Autocorrelaci√≥n Simple')
plt.title('Funci√≥n de Autocorrelaci√≥n Simple')
plt.show()

# Calcular la funci√≥n de autocorrelaci√≥n parcial
plot_pacf(residuos['residuos'], lags=10)
plt.xlabel('Lags')
plt.ylabel('Autocorrelaci√≥n Parcial')
plt.title('Funci√≥n de Autocorrelaci√≥n Parcial')
plt.show()
```

El error del mdoelo es ruido blanco y con distribuci√≥n nomral
